# City by City Analysis

## Riyadh - Throughput

```python
#Merging CSVs - Download
df0 = pd.read_csv(r'riyadh\riyadh_download_0.csv', parse_dates=['date'], index_col='date')
df0 = df0[~df0.index.month.isin([1, 2, 3, 4, 5])]
df1 = pd.read_csv(r'riyadh\riyadh_download_1.csv', parse_dates=['date'], index_col='date')
df2 = pd.read_csv(r'riyadh\riyadh_download_2.csv', parse_dates=['date'], index_col='date')
df3 = pd.read_csv(r'riyadh\riyadh_download_3.csv', parse_dates=['date'], index_col='date')
df4 = pd.read_csv(r'riyadh\riyadh_download_4.csv', parse_dates=['date'], index_col='date')
df5 = pd.read_csv(r'riyadh\riyadh_download_5.csv', parse_dates=['date'], index_col='date')
df6 = pd.read_csv(r'riyadh\riyadh_download_6.csv', parse_dates=['date'], index_col='date')
dfs_D = [df0, df1, df2, df3, df4, df5, df6]
dfR = pd.concat(dfs_D)
riyadh = dfR.resample('M').agg({'MeanThroughputMbps':'mean'}).rename(columns={"MeanThroughputMbps":"Throughput_Riyadh"})
riyadh_by_year = riyadh.copy()
riyadh_by_year['month'] = pd.to_datetime(riyadh_by_year.index).month
riyadh_by_year['year'] = pd.to_datetime(riyadh_by_year.index).year
Riyadh = dict(tuple(riyadh_by_year.groupby('year'))) # for all cities plot

#Merging CSVs - Upload
df0 = pd.read_csv(r'riyadh\riyadh_upload_0.csv', parse_dates=['date'], index_col='date')
df0 = df0[~df0.index.month.isin([1, 2, 3, 4, 5])]
df1 = pd.read_csv(r'riyadh\riyadh_upload_1.csv', parse_dates=['date'], index_col='date')
df2 = pd.read_csv(r'riyadh\riyadh_upload_2.csv', parse_dates=['date'], index_col='date')
df3 = pd.read_csv(r'riyadh\riyadh_upload_3.csv', parse_dates=['date'], index_col='date')
df4 = pd.read_csv(r'riyadh\riyadh_upload_4.csv', parse_dates=['date'], index_col='date')
df5 = pd.read_csv(r'riyadh\riyadh_upload_5.csv', parse_dates=['date'], index_col='date')
df6 = pd.read_csv(r'riyadh\riyadh_upload_6.csv', parse_dates=['date'], index_col='date')
dfs_D = [df0, df1, df2, df3, df4, df5, df6]
dfRU = pd.concat(dfs_D)
riyadhU = dfRU.resample('M').agg({'MeanThroughputMbps':'mean'}).rename(columns={"MeanThroughputMbps":"Throughput_Riyadh"})
riyadhU_by_year = riyadhU.copy()
riyadhU_by_year['month'] = pd.to_datetime(riyadhU_by_year.index).month
riyadhU_by_year['year'] = pd.to_datetime(riyadhU_by_year.index).year
RiyadhU = dict(tuple(riyadhU_by_year.groupby('year'))) # for all cities plot
```

## Riyadh Plot

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(nrows=2, ncols=2)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])

#Five-Year Data
ax1.plot(riyadh, linestyle='--', marker='o', label='Download')
ax1.plot(riyadhU, marker='X', label='Upload')
#sb.lineplot(data=riyadh, x=riyadh.index,y='Throughput_Riyadh', marker='o', palette=palette, ax = ax1, legend = True)
#sb.lineplot(data=riyadhU, x=riyadhU.index,y='Throughput_Riyadh', marker='X', palette=palette, ax = ax1)
ax1.set_title("Average Internet Speeds Recorded Between June 2018 and May 2023 in Riyadh, Saudi Arabia\n\n(A) Download and Upload", fontdict={'fontsize': 20})
ax1.set_xticks(['2018-06-30', '2018-12-31', '2019-12-31', '2020-12-31', '2021-12-31', '2022-12-31', '2023-05-31'])
ax1.set_xticklabels(['Jun\n2018', 'Dec\n2018', 'Dec\n2019','Dec\n2020', 'Dec\n2021', 'Dec\n2022', 'May\n2023'])
ax1.tick_params(axis = 'both', which = 'both', labelsize = 15)
#ax1.tick_params(axis = 'both', which = 'minor', labelsize = 15)

ax1.set_xlabel('Five-Year Period', fontdict={'fontsize': 17}, labelpad = 15)
ax1.set_ylabel('Throughput (Mbps)', fontdict={'fontsize': 17}, labelpad = 15)
ax1.legend(prop={'size': 15})

#Year-by_Year 

sb.lineplot(x='month',y='Throughput_Riyadh',data=riyadh_by_year, hue = 'year', 
            style='year', markers=True, palette=palette2, size='year', 
            legend=False, ax = ax2) # for download

sb.lineplot(x='month',y='Throughput_Riyadh',data=riyadhU_by_year, hue = 'year', 
            style='year', markers=True, palette=palette2, size='year', 
            legend=False, ax = ax3) # for upload

ax2.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax2.set_xlabel('Months of the Year', fontdict={'fontsize': 17}, labelpad = 15)
ax2.set_ylabel('Throughput (Mbps)', fontdict={'fontsize': 17}, labelpad = 15)
ax2.set_title('(B) Download', fontdict={'fontsize': 20})

ax3.tick_params(axis = 'both', which = 'major', labelsize = 15)
ax3.set_xlabel('Months of the Year', fontdict={'fontsize': 17}, labelpad = 15)
ax3.set_ylabel('Throughput (Mbps)', fontdict={'fontsize': 17}, labelpad = 15)
ax3.set_title('(C) Upload', fontdict={'fontsize': 20})
ax3.yaxis.tick_right()
ax3.yaxis.set_label_position('right')


ax2.legend(labels=['2018', '2019','2020', '2021', '2022', '2023'], prop={'size': 16}, bbox_to_anchor=(1.3, 1))
plt.savefig('figures/riyadhthru.pdf')
```

```python
#calcuate percentage increase
riyadh['Perc_Increase'] = None
years = riyadh.index.year.unique()
for year in years:
    # If it's the first year in the dataset, calculate from June to December
    if year == 2018:
        start = riyadh.loc['2018-06-30', 'Throughput_Riyadh']
        end = riyadh.loc['2018-12-31', 'Throughput_Riyadh']
        change = ((end - start) / start) * 100
        riyadh.loc['2018-12-31', 'Perc_Increase'] = round(change)
    # If it's the last year in the dataset, calculate from January to May
    elif year == 2023:
        start = riyadh.loc['2023-01-31', 'Throughput_Riyadh']
        end = riyadh.loc['2023-05-31', 'Throughput_Riyadh']
        change = ((end - start) / start) * 100
        riyadh.loc['2023-05-31', 'Perc_Increase'] = round(change)
    # For other years, calculate from January to December
    else:
        start = riyadh.loc[f'{year}-01-31', 'Throughput_Riyadh']
        end = riyadh.loc[f'{year}-12-31', 'Throughput_Riyadh']
        change = ((end - start) / start) * 100
        riyadh.loc[f'{year}-12-31', 'Perc_Increase'] = round(change)

#riyadh
# Calculate the average Throughput_Riyadh for 2018 (from June to December) and 2023 (January to May)
avg_2018 = riyadh.loc['2018-06':'2018-12', 'Throughput_Riyadh'].mean()
avg_2023 = riyadh.loc['2023-01': '2023-05', 'Throughput_Riyadh'].mean()

# Calculate the percentage change from 2018 to 2023
change_2018_to_2023 = ((avg_2023 - avg_2018) / avg_2018) * 100
print('Average 2018:', round(avg_2018, 2), '  ', 'Average 2023:', round(avg_2023, 2))
print('The percentage change from 2018 to 2023 is:', round(change_2018_to_2023))

# Filter the DataFrame to keep only June and December of 2018, and January and the last month of other years
filtered_riyadh = riyadh.loc['2018-06'].append(riyadh[(riyadh.index.month == 1) | (riyadh.index.month == 12)])
filtered_riyadh = filtered_riyadh.append(riyadh.loc['2023-5'])
# For the last year, we want to keep the data for May instead of December
#last_month_2023 = riyadh.loc['2023'].index.month.max()
#riyadh #= riyadh.append(riyadh.loc[f'2023-{last_month_2023}'])
#last_month_2023
filtered_riyadh['Throughput_Riyadh'] = filtered_riyadh['Throughput_Riyadh'].round(2)
filtered_riyadh
```

```python
fig, ax = plt.subplots(3,2, figsize = (20,12))
fig.suptitle('Year by Year Comparison of Download Speeds of the Six Major GCC cities\n\n', fontsize = 20)

cities = {'Riyadh': Riyadh, 'Dubai': Dubai, 'Kuwait': Kuwait, 'Doha': Doha, 'Muscat': Muscat, 'Manama': Manama}
styles = {'Riyadh': {"linestyle": "solid", "marker": 'X'}, 
          'Dubai': {"linestyle": "dotted", "marker": 'o'},
          'Kuwait': {"linestyle": "dashed", "marker": '>'}, 
          'Doha': {"linestyle": "dashdot", "marker": 'd'}, 
          'Muscat': {"linestyle": (0, (1, 10)), "marker": 's'}, 
          'Manama': {"linestyle": (0, (3, 1, 1, 1, 1, 1)), "marker": 'P'}}
years = [2018, 2019, 2020, 2021, 2022, 2023]

for i, year in enumerate(years):
    for city, df in cities.items():
        sb.lineplot(x='month', 
                    y=f'Throughput_{city}',
                    data=df[year], 
                    markers=True, 
                    linestyle = styles[city]['linestyle'], 
                    linewidth = 2.5, 
                    marker = styles[city]['marker'], 
                    legend=False, 
                    ax = ax[i//2, i%2])
    ax[i//2, i%2].set_title(str(year), fontdict={'fontsize': 20})
    ax[i//2, i%2].set_xlabel('')
    ax[i//2, i%2].set_ylabel('Throughput (Mbps)', fontdict={'fontsize': 15}, labelpad = 15)
    ax[i//2, i%2].tick_params(axis = 'both', which = 'major', labelsize = 15)
    
    if (i%2 != 0):
        ax[i//2, i%2].yaxis.tick_right()
        ax[i//2, i%2].yaxis.set_label_position('right')

ax[2,0].set_xlabel('Months', fontdict={'fontsize': 15}, labelpad = 15)
ax[2,1].set_xlabel('Months', fontdict={'fontsize': 15}, labelpad = 15)
ax[2,1].set_xticks([1,2,3,4,5])
ax[0,0].legend(labels=['Riyadh','Dubai', 'Kuwait City', 'Doha', 'Muscat', 'Manama'], 
               prop={'size': 16}, bbox_to_anchor=(1.3, 1.2))

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.25, 
                    hspace=0.3)

plt.savefig('figures/all_download_thru1.pdf', bbox_inches='tight')
```
