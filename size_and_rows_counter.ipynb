```python
import os

def count_csv_rows(directory, file_prefix):
    count = 0
    for filename in os.listdir(directory):
        if filename.startswith(file_prefix) and filename.endswith('.csv'):
            with open(os.path.join(directory, filename), 'r', encoding='utf-8-sig') as file:
                next(file)  # Skip the header row
                for _ in file:
                    count += 1
    return count

current_dir = os.getcwd()
cities = ['riyadh', 'dubai', 'kuwaitcity', 'doha', 'muscat', 'manama']

total_downloads_count = 0
total_uploads_count = 0

for city in cities:
    base_path = os.path.join(current_dir, city)
    
    downloads_count = count_csv_rows(base_path, f'{city}_download_')
    uploads_count = count_csv_rows(base_path, f'{city}_upload_')

    total_downloads_count += downloads_count
    total_uploads_count += uploads_count

    print(f"The number of rows in {city.capitalize()} Downloads is {downloads_count}.")
    print(f"The number of rows in {city.capitalize()} Uploads is {uploads_count}.")

print(f"Total number of downloads across all cities: {total_downloads_count}")
print(f"Total number of uploads across all cities: {total_uploads_count}")
print(f"Total number of downloads and uploads combined: {total_downloads_count + total_uploads_count}")
```

```python
import os

def count_csv_rows(directory, file_prefix):
    count = 0
    for filename in os.listdir(directory):
        if filename.startswith(file_prefix) and filename.endswith('.csv'):
            with open(os.path.join(directory, filename), 'r', encoding='utf-8-sig') as file:
                next(file)  # Skip the header row
                for _ in file:
                    count += 1
    return count

def calculate_total_file_size(directory, file_prefix):
    total_size = 0
    for filename in os.listdir(directory):
        if filename.startswith(file_prefix) and filename.endswith('.csv'):
            total_size += os.path.getsize(os.path.join(directory, filename))
    return total_size

current_dir = os.getcwd()
cities = ['riyadh', 'dubai', 'kuwaitcity', 'doha', 'muscat', 'manama']

total_downloads_count = 0
total_uploads_count = 0
total_files_size = 0

for city in cities:
    base_path = os.path.join(current_dir, city)
    
    downloads_count = count_csv_rows(base_path, f'{city}_download_')
    uploads_count = count_csv_rows(base_path, f'{city}_upload_')

    total_downloads_count += downloads_count
    total_uploads_count += uploads_count
    
    downloads_size = calculate_total_file_size(base_path, f'{city}_download_')
    uploads_size = calculate_total_file_size(base_path, f'{city}_upload_')
    city_files_size = downloads_size + uploads_size
    total_files_size += city_files_size

    print(f"The number of rows in {city.capitalize()} Downloads is {downloads_count}.")
    print(f"The number of rows in {city.capitalize()} Uploads is {uploads_count}.")
    print(f"Total file size for {city.capitalize()} (downloads and uploads): {city_files_size} bytes.")

print(f"Total number of downloads across all cities: {total_downloads_count}")
print(f"Total number of uploads across all cities: {total_uploads_count}")
print(f"Total number of downloads and uploads combined: {total_downloads_count + total_uploads_count}")
print(f"Total file size for all cities (downloads and uploads): {round(total_files_size/(1024**2), 2)} megabytes.")
```
